```python
import numpy as np
import pandas as pd
import pickle

with open('api.pkl', 'rb') as file:
    data = pickle.load(file)

print(data)
```

```python
for key, value in data.items():
    print(f"{key}: {value}")
    break
```

```python
df = {
    "classes":[],
    "meainings":[]
}
for key,val in data.items():
    real_val = ''
    for s in val:
        if s == '':
            continue
        else:
            real_val+=s
    df['classes'].append(key)
    df['meainings'].append(real_val)
for key, value in data.items():
    print(f"{key}: {value}")
    break
```

```python
df
```

```python
df = pd.DataFrame(df)
```

```python
df.to_csv('csv_files\def_api.csv',index=False)
```
